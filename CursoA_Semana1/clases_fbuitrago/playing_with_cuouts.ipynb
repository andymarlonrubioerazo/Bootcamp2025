```python
import numpy as np
from astropy.io import fits
import os
import sys
import pdb
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
```

```python
path_image = "./"
image_name = "17058_goodss_I_hudf.fits"
zeropoint_I = 25.94
pixel_scale = 0.03 #[arcsec/pix]
max_sb = 18 #mag/arcsec^2
min_sb = 32 #mag/arcsec^2
```

```python
hdu = fits.open(path_image + image_name)
img = hdu[0].data
hdr = hdu[0].header
```

```python
plt.imshow(img, origin='lower', interpolation="none")
```

```python
plt.imshow(np.log10(img), origin='lower', interpolation="none")
```

```python
plt.imshow(-2.5*np.log10(img)+zeropoint_I+(5*np.log10(pixel_scale)), origin='lower', interpolation="none", cmap="viridis_r")
```

```python
imshow_img = plt.imshow(-2.5*np.log10(img)+zeropoint_I+(5*np.log10(pixel_scale)), origin='lower', interpolation="none", cmap="viridis_r", vmin=max_sb, vmax=min_sb)
cb = plt.colorbar(imshow_img)
cb.set_label(r'Surface brightness [mag/arcsec$^{2}$]')

scalebar = AnchoredSizeBar(plt.gca().transData, 10./pixel_scale, "10 arcsec", borderpad=1.0, sep=5, loc="lower right", pad=0.1, color="blue", frameon=True, size_vertical=3)
plt.gca().add_artist(scalebar)
```
